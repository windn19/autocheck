### Задача 2

Требуется функцию для преобразования координат из декартовой системы в полярную.

***Справка***:

* В декартовой системе точка описывается с помощью двух координат (x,y)
* В полярной системе точка описывается с помощью двух координат (полярный угол и полярный радиус)

Для перехода из декартовой системы в полярную необходимо воспользоваться следующими формулами:

- Полярный радиус: $r=\sqrt{x^2 + y^2}$
- Полярный угол: $tg(a)=\frac{y}{x}$, следовательно сам угол можем вычислить следующим образом: $a = arctan(\frac{y}{x})$, где **arctan** - функция арктангенс, которая реализована в библиотеке **numpy** в виде метода `numpy.arctan()`.
Важно! Функция возвращает значение от -pi/2 до pi/2, поэтому требуется доп. условие: x>0, чтобы верно перевести только с помощью этой функции точки с отрицательными x в полярные координаты. При x < 0 угол стоит считать следующим образом: np.arctan(y/x)+np.pi

Функция на вход кортеж из двух координат, а возвращает из кортеж из двух значений (полярный радиус, полярный угол).
 -  Функция должна называться **find_angle**
 -  на вход она принимает кортеж из двух значений - (4, 3)
 -  возвращает кортеж из двух значений - (5.0, 0.6435011087932844)

```
from math import sqrt

def find_angle(ang):
  x, y = ang
  r = sqrt(x ** 2 + y ** 2)
  a = np.arctan(y/x) if x > 0 else np.arctan(y/x) + np.pi
  return r, a

find_angle((-2, 6))
```
